```python
NER_LIST = []
```

#Spacy

```python
import spacy
from spacy import displacy

NER = spacy.load("en_core_web_sm")
```

```python
cnt = 1
for example in df.directions:
  temp_list = []
  ner_results = NER(example)
  #print(cnt, " : ", example)
  cnt = cnt + 1
  for word in ner_results.ents:
    temp_list.append(str(word.text))
    #print("NER ======> ", word.text)
  NER_LIST.append(temp_list)
```

#NLTK

```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
 
```

```python
cnt = 0
for sentence in df.directions:
  #print(cnt, ":", sentence)
  for sent in nltk.sent_tokenize(sentence):
    for chunk in nltk.ne_chunk(nltk.pos_tag(nltk.word_tokenize(sent))):
      if hasattr(chunk, 'label'):
          NER_LIST[cnt].append(' '.join(c[0] for c in chunk))
          #print("NER ====> ", ' '.join(c[0] for c in chunk))
  cnt = cnt + 1
```

#RecipeNLG merge

```python
punc = '''!()-[]{};:'"\<>./?@#$%^&*_~'''

def clean_text(txt):
 
# Removing punctuations in string
# Using loop + punctuation string
  for ele in txt:
      if ele in punc:
          txt = txt.replace(ele, "")
  return txt.split(",")
```

```python
idx = 0
for i in df.NER:
  txt = clean_text(i)
  for j in txt:
    NER_LIST[idx].append(j.strip())
  idx = idx + 1
```

```python
for i in range(len(NER_LIST)):
  NER_LIST[i] = str(list(set(NER_LIST[i])))
```

```python
df["Extended_NER"] = NER_LIST
```

```python
df = df[['title', 'directions', 'NER', 'genre', 'label', 'Extended_NER']]
```
